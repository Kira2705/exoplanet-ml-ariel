```python
import panda as pd
import numpy as np
import matplotlib.pyplot as plt

test_airs_ch0 = pd.read_parquet('/kaggle/input/ariel-data-challenge-2024/test/499191466/AIRS-CH0_calibration')
test_airs_signal = pd.read_parquet('/kaggle/input/ariel-data-challenge-2024/test/499191466/AIRS-CH0_signal.parquet')
test_fgs1_calibration = pd.read_parquet('/kaggle/input/ariel-data-challenge-2024/test/499191466/FGS1_calibration')
test_fgs1_signal = pd.read_parquet('/kaggle/input/ariel-data-challenge-2024/test/499191466/FGS1_signal.parquet')

print("AIRS-CH0 Siganal shape:", test_airs_signal.shape)
print(test_airs_signal.head())

print("\n" + "="*50)
print("FGS1 Signal shape:", test_fgs1_signal.shape)
print(test_fgs1_signal.head())
```

```python
# Plot some example spectra
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(test_airs_signal.iloc[0])
plt.title('AIRS-CH0 Signal - First Observation')
plt.xlabel('Wavelength Index')
plt.ylabel('Signal')

plt.subplot(1, 2, 2)
plt.plot(test_fgs1_signal.iloc[0])
plt.title('FGS1 Signal - First Observation')
plt.xlabel('Wavelength Index')
plt.ylabel('Signal')

plt.tight_layout()
plt.show()
```

```python
# Check what's in each file
print("AIRS-CH0 Signal columns:")
print(test_airs_signal.columns.tolist())

print("\nFGS1 Signal columns:")
print(test_fgs1_signal.columns.tolist())

# Basic statistics
print("\nAIRS-CH0 Signal statistics:")
print(test_airs_signal.describe())
```

```python
# Let's zoom into the interesting region
plt.figure(figsize=(12, 5))
plt.plot(test_airs_signal.iloc[0])
plt.axhline(y=500, color='r', linestyle='--', label='Baseline')
plt.xlim(5000, 7000)  # Zoom into the peak region
plt.xlabel('Wavelength Index')
plt.ylabel('Signal')
plt.title('AIRS-CH0 Signal - Zoomed into Peak Region')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Visualize AIRS-CH0 calibration
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(test_airs_ch0.iloc[0])
plt.title('AIRS-CH0 Calibration - First Row')
plt.xlabel('Column Index')
plt.ylabel('Calibration Value')
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 2)
plt.plot(test_fgs1_calibration.iloc[0])
plt.title('FGS1 Calibration - First Row')
plt.xlabel('Column Index')
plt.ylabel('Calibration Value')
plt.grid(True, alpha=0.3)

# If there are multiple rows, show them all
plt.subplot(2, 2, 3)
for i in range(min(5, len(test_airs_ch0))):
    plt.plot(test_airs_ch0.iloc[i], alpha=0.7, label=f'Row {i}')
plt.title('AIRS-CH0 Calibration - Multiple Rows')
plt.xlabel('Column Index')
plt.ylabel('Calibration Value')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(2, 2, 4)
for i in range(min(5, len(test_fgs1_calibration))):
    plt.plot(test_fgs1_calibration.iloc[i], alpha=0.7, label=f'Row {i}')
plt.title('FGS1 Calibration - Multiple Rows')
plt.xlabel('Column Index')
plt.ylabel('Calibration Value')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print info about the data
print("AIRS-CH0 Calibration Info:")
print(f"Shape: {test_airs_ch0.shape}")
print(f"Value range: {test_airs_ch0.min().min():.4f} to {test_airs_ch0.max().max():.4f}")
print(f"\nFirst few values:\n{test_airs_ch0.iloc[0, :10]}")

print("\n" + "="*60)
print("FGS1 Calibration Info:")
print(f"Shape: {test_fgs1_calibration.shape}")
print(f"Value range: {test_fgs1_calibration.min().min():.4f} to {test_fgs1_calibration.max().max():.4f}")
print(f"\nFirst few values:\n{test_fgs1_calibration.iloc[0, :10]}")
```
